# Medium level

### Test case 2 : write a function to count vowels in a string

Count vowels: a, i, u, e, o (case-insensitive).

```python
# def ...(text):
#     return ...

# Test
# print(...("Hello World"))  # Expected: 3  (e,o,o)
# print(...("PYTHON"))       # Expected: 1  (o)

def cnt_vwl(text):
    a = text.lower()
    cnt = 0
    box = {'a','e','i','o','u'}
    b = []
    for word in a:
        if word in box:
            cnt += 1
            b.append(word)
    
    c = ",".join(b)
    return f'{cnt} ({c})'

print(cnt_vwl("Hello World"))
print(cnt_vwl("PYTHON"))
```

# Medium level
### Test case 3 : write a function to summarize shopping cart

Given a list of prices, return (total, average, max). Use built-in functions when possible.

```python
# def ...(prices):
#     return ...

# Test
# print(...([10, 20, 30]))         # Expected: (60, 20.0, 30)
# print(...([5]))                  # Expected: (5, 5.0, 5)

def sum(prices):
    total = 0
    avg = 0
    max = 0
    
    for i in prices:
        total += i
        if i > max:
            max = i
        
    avg = float(total / len(prices))
    return f"({total}, {avg}, {max})"

print(sum([10,20,30]))
print(sum([5]))
```

# Hard level
### Test case 4 : write a function to remove duplicates while keeping order

Example: `[1,2,1,3,2] -> [1,2,3]`

```python
# def ...(items):
#     return ...

# Test
# print(...([1, 2, 1, 3, 2]))                 # Expected: [1, 2, 3]
# print(...(["a", "b", "a", "c", "b"]))       # Expected: ['a', 'b', 'c']

def rm_dup(items):
    box = set()
    for i in items:
        box.add(i)
        
    return sorted(box)

print (rm_dup([1,2,1,3,2]))
print(rm_dup(["a", "b", "a", "c", "b"]))
```

### Test case 5 : create a class StudentGradeBook

Requirements:
- `__init__(self, name)` sets name and empty list of scores
- `add_score(self, score)` adds score (0-100). If invalid, return "Invalid score"
- `average(self)` returns average score (float). If no scores, return 0
- `status(self)` returns:
  - "Pass" if average >= 75
  - "Remedial" if 60 <= average < 75
  - "Fail" if average < 60


```python
class StudentGradeBook:
    def __init__(self, name):
        self.name = name
        self.score = []

    def add_score(self, score):
        if score < 0 or score > 100:
            return "Invalid score"
        self.score.append(score)
        return f"Added {score}"

    def average(self):
        if len(self.score) == 0:
            return 0
        tot = 0
        for i in self.score:
            tot += i
        return float(tot/len(self.score))

    def status(self):
        avg = self.average()
        if avg >= 75:
            return 'Pass'
        elif avg < 75 and avg >= 60:
            return "Remedial"
        else:
            return "Fail"
            
        

# Test
s = StudentGradeBook("Alya")
print(s.average())              
print(s.status())               
print(s.add_score(80))         
print(s.add_score(70))         
print(s.add_score(120))        
print(s.average())              
print(s.status())               

```
