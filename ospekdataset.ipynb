```python
# Parse dates
for col in ["INSPECTION DATE", "GRADE DATE", "RECORD DATE"]:
    df[col] = pd.to_datetime(df[col], errors="coerce")

# Filter 3 tahun terakhir (berdasarkan snapshot RECORD DATE)
record_date = df["RECORD DATE"].dropna().max()
start_date = record_date - pd.DateOffset(years=3)

df_recent = df[
    (df["INSPECTION DATE"].notna()) &
    (df["INSPECTION DATE"] >= start_date) &
    (df["INSPECTION DATE"] <= record_date)
].copy()

# Bikin inspection-level table (hindari duplikasi per violation)
inspection_key = ["CAMIS", "INSPECTION DATE", "INSPECTION TYPE"]
df_ins = df_recent.drop_duplicates(subset=inspection_key).copy()
df_ins["SCORE"] = pd.to_numeric(df_ins["SCORE"], errors="coerce")

df_ins["BORO"] = df_ins["BORO"].replace("0", "Unknown")

has_critical = (
    df_recent.groupby(inspection_key)["CRITICAL FLAG"]
    .apply(lambda s: (s == "Critical").any())
    .reset_index(name="has_critical")
)

df_ins = df_ins.merge(has_critical, on=inspection_key, how="left")
df_ins["has_critical"] = df_ins["has_critical"].fillna(False)

df_ins.info()
```

```python
df_ins["month"] = df_ins["INSPECTION DATE"].dt.to_period("M").dt.to_timestamp()

graded = df_ins[df_ins["GRADE"].isin(["A","B","C"])].copy()
pctA_month = graded.groupby("month")["GRADE"].apply(lambda s: (s=="A").mean()*100)

plt.figure(figsize=(10,4))
plt.plot(pctA_month.index, pctA_month.values, marker="o")
plt.title("% Grade A per Bulan (3 tahun terakhir)")
plt.xlabel("Bulan")
plt.ylabel("% Grade A")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top violation codes (Overall vs Critical) - violation level (1 chart)
df_recent["VIOLATION CODE"] = df_recent["VIOLATION CODE"].astype(str)
overall_counts = df_recent["VIOLATION CODE"].value_counts(dropna=False)
critical_counts = df_recent.loc[df_recent["CRITICAL FLAG"]=="Critical","VIOLATION CODE"].value_counts(dropna=False)

top_codes = overall_counts.head(15).index.tolist()
plot_df = pd.DataFrame({
    "overall": overall_counts.reindex(top_codes).fillna(0).astype(int),
    "critical": critical_counts.reindex(top_codes).fillna(0).astype(int)
}, index=top_codes)

x = np.arange(len(top_codes))
width = 0.42

plt.figure(figsize=(12,5))
plt.bar(x - width/2, plot_df["overall"].values, width, label="Overall")
plt.bar(x + width/2, plot_df["critical"].values, width, label="Critical")
plt.title("Top 15 Violation Codes: Overall vs Critical (3 tahun terakhir)")
plt.xlabel("Violation Code")
plt.ylabel("Jumlah baris pelanggaran")
plt.xticks(x, top_codes, rotation=45, ha="right")
plt.legend()
plt.show()
```

```python
# Rata-rata SCORE per BORO
boro_df = df_ins[(df_ins["BORO"].notna()) & (df_ins["BORO"] != "0")].copy()
boro_stats = boro_df.groupby("BORO")["SCORE"].mean().sort_values()
plt.figure()
plt.bar(boro_stats.index, boro_stats.values)
plt.title("Rata-rata Score per BORO - 3 tahun terakhir")
plt.xlabel("BORO")
plt.ylabel("Mean Score")
plt.show()
```

```python
# Rata-rata SCORE untuk 15 cuisine terpopuler
cuisine_df = df_ins.copy()
cuisine_df["CUISINE DESCRIPTION"] = cuisine_df["CUISINE DESCRIPTION"].fillna("Unknown")
top_cuisines = cuisine_df["CUISINE DESCRIPTION"].value_counts().head(15).index
cuisine_stats = (
    cuisine_df[cuisine_df["CUISINE DESCRIPTION"].isin(top_cuisines)]
    .groupby("CUISINE DESCRIPTION")["SCORE"]
    .mean()
    .sort_values()
)

plt.figure(figsize=(12,5))
plt.bar(cuisine_stats.index, cuisine_stats.values)
plt.title("Rata-rata Score: 15 Cuisine Terpopuler (3 tahun terakhir)")
plt.xlabel("Cuisine Description")
plt.ylabel("Mean Score")
plt.xticks(rotation=45, ha="right")
```

```python
# Repeat offenders (bucket jumlah inspeksi yang punya critical violation per restoran)
inspection_has_critical = (
    df_recent.groupby(inspection_key)["CRITICAL FLAG"]
    .apply(lambda s: (s=="Critical").any())
    .reset_index(name="has_critical")
)

crit_counts_per_rest = inspection_has_critical.groupby("CAMIS")["has_critical"].sum().astype(int)
bucket = pd.cut(crit_counts_per_rest, bins=[-0.1,0,1,2,3,10_000], labels=["0","1","2","3","4+"])
bucket_counts = bucket.value_counts().reindex(["0","1","2","3","4+"])

plt.figure()
plt.bar(bucket_counts.index.astype(str), bucket_counts.values)
plt.title("Repeat Offenders: # inspeksi critical per restoran (3 tahun terakhir)")
plt.xlabel("Jumlah inspeksi critical (bucket)")
plt.ylabel("Jumlah restoran")
plt.show()
```
