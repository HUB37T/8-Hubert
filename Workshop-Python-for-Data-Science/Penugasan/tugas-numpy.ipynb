## **📌 Soal 4: Boolean Masking & `np.where()`**

Diberikan array suhu berikut:


```python
temps = np.array([29, 31, 35, 28, 33, 27, 30, 36, 32, 26])
print(temps)
```

Kerjakan hal berikut:
1. Ambil semua suhu yang **lebih dari 30**
2. Hitung berapa banyak suhu yang **lebih dari atau sama dengan 32**
3. Buat array label baru bernama `status` dengan aturan:
   - `"panas"` jika suhu > 32
   - `"normal"` jika suhu 29 s.d. 32
   - `"dingin"` jika suhu < 29

**Hint:** untuk poin (3), kamu bisa menggunakan `np.where()` bertingkat.


```python
# Tulis jawabanmu di sini
bool_satu = temps > 30
bool_dua = temps >= 32

arr_status = np.where(temps > 32, "Panas", np.where(temps >= 29, "Normal", "Dingin"))
print("1. Ambil semua suhu yang lebih dari 30:\n", temps[bool_satu])
print("\n2. Hitung berapa banyak suku yang lebih dari atau sama dengan 32:\n", np.sum(bool_dua))
print("\n3. Buat array label baru dengan nama status dengan aturan:\n",arr_status)
```

## **📌 Soal 5: View vs Copy**

Buat array berikut:


```python
x = np.array([10, 20, 30, 40, 50, 60])
print(x)
```

Lalu lakukan:
1. Buat `view_part` dari elemen indeks 1 sampai 4 menggunakan slicing
2. Buat `copy_part` dari elemen indeks `[1, 2, 3, 4]` menggunakan advanced indexing
3. Ubah elemen pertama `view_part` menjadi `999`
4. Ubah elemen pertama `copy_part` menjadi `-100`
5. Tampilkan `x`, `view_part`, dan `copy_part`

Terakhir, tuliskan dengan singkat:
- mana yang menyebabkan array asli berubah?
- mana yang tidak?


```python
# Tulis jawabanmu di sini
# Tambahkan penjelasan singkat dalam komentar juga.
view_part = x[1:5]
copy_part = x[[1, 2, 3, 4]]
view_part[0] = 999
copy_part[0] = -100
print("Array x:\n", x)
print("\nArray View:\n", view_part)
print("\nArray Copy:\n", copy_part)
```

1. Yang menyebabkan array asli berubah adalah metode slicing karena metode slicing menggunakan reference dari array asli sehingga jika array copyan diubah maka array asli juga ikut berubah
2. Yang tidak menyebabkan array asli berubah adalah metode advanced indexing karena yang diambil hanya isi arraynya bukan referencenya

## **📌 Soal 6: Statistik Dasar & Missing Value (`NaN`)**

Gunakan array berikut:


```python
values = np.array([12, 15, np.nan, 18, 21, np.nan, 24, 30])
print(values)
```

Tampilkan:
1. Rata-rata biasa dengan `np.mean()`
2. Rata-rata yang mengabaikan `NaN` dengan `np.nanmean()`
3. Standar deviasi yang mengabaikan `NaN` dengan `np.nanstd()`
4. Nilai minimum dan maksimum yang mengabaikan `NaN`

Lalu jelaskan singkat kenapa hasil `np.mean()` bisa berbeda dari `np.nanmean()`.


```python
# Tulis jawabanmu di sini
print("1. Rata-rata biasa dengan np.mean:\n", np.mean(values))
print("\n2. Rata-rata dengan np.nanmean (mengabaikan NaN):\n", np.nanmean(values))
print("\n3. Standar deviasi yang mengabaikan NaN dengan np.nanstd:\n", np.nanstd(values))
print("\n4. Nilai minimum dan maksimum yang mengabaikan Nan:\n", np.nanmin(values), np.nanmax(values))
```

**Penjelasan np.mean() dan np.nanmean()**

np.mean() dan np.nanmean() berbeda karena jika di dalam array terdapat nilai NaN dan kita menggunakan np.mean() biasa, maka hasil rata-ratanya akan menjadi nan sedangkan jika menggunakan np.nanmean() nilai NaN pada array tidak akan diikutkan pada proses perhitungan rata-rata nantinya sehingga bisa menghasilkan nilai rata-rata dalam bentuk numerik

## **📌 Soal 7: Broadcasting untuk Normalisasi**

Diberikan matriks fitur berikut:


```python
X = np.array([
    [170, 65, 28],
    [165, 55, 24],
    [180, 75, 30],
    [175, 70, 29]
], dtype=float)

print(X)
```

Anggap setiap kolom merepresentasikan fitur yang berbeda.  
Lakukan normalisasi z-score sederhana dengan rumus:

$$
Z = \frac{X - \text{mean kolom}}{\text{std kolom}}
$$

Tampilkan:
1. Mean per kolom
2. Std per kolom
3. Hasil normalisasi `Z`

**Hint:** gunakan `axis=0` dan manfaatkan broadcasting NumPy.


```python
# Tulis jawabanmu di sini
arr_mean = np.mean(X, axis=0)
arr_std = np.std(X, axis=0)
X_normalized = (X - arr_mean) / arr_std
print("\nMean per kolom:\n", arr_mean)
print("\nStandard Deviation per kolom:\n", arr_std)
print("\nArray Normalized:\n", X_normalized)
```

## **📌 Soal 8: Reshape & Mini Challenge**

Gunakan array 1 dimensi berikut:


```python
z = np.arange(1, 25)
print(z)
```

Kerjakan hal berikut:
1. Ubah `z` menjadi matriks berukuran `6x4`
2. Ubah hasilnya menjadi bentuk `3x2x4`
3. Ambil kembali seluruh data tersebut menjadi array 1 dimensi
4. Bonus: buat array window 4 elemen yang bergeser dari `z`, sehingga bentuk akhirnya menjadi:

`[[1,2,3,4], [2,3,4,5], ..., [21,22,23,24]]`

**Hint bonus:** boleh gunakan loop **atau** `sliding_window_view` jika kamu sudah tahu.


```python
# Tulis jawabanmu di sini
arr_hsl = z.reshape(6,4)
arr_dua = arr_hsl.reshape(3,2,4)
arr_tiga = arr_dua.reshape(1,24)
arr_empat = np.lib.stride_tricks.sliding_window_view(z, window_shape=4)
print("1. Ubah z menjadi matriks berukuran 6x4:\n", arr_hsl)
print("\n2. Ambil hasilnya menjadi bentuk 3x2x4:\n", arr_dua)
print("\n3. Ambil kembali seluruh data tersebut menjadi array 1 dimensi:\n", arr_tiga)
print("\n4. Buat array window 4 elemen yang bergeser dari z:\n", arr_empat.tolist())
```
