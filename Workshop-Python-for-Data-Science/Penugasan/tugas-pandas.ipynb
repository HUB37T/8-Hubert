## **📦 Import Library & Load Dataset**

Dataset yang digunakan adalah **`taxis.csv`**.

Sumber dataset:  
`https://raw.githubusercontent.com/mwaskom/seaborn-data/master/taxis.csv`


```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/taxis.csv"
df = pd.read_csv(url)

df.head()
```

## **📌 Soal 1: Kenali Dataset Terlebih Dahulu**

Tampilkan informasi berikut dari dataset `df`:
1. 5 baris teratas
2. Shape dataset
3. Nama semua kolom
4. Informasi detail data dengan `info()`
5. Jumlah missing value per kolom

**Tujuan soal:** membiasakan diri membaca struktur dataset sebelum mulai analisis.


```python
# Tulis jawabanmu di sini
print("1. 5 baris teratas:\n", df.head())
print("\n2. Shape dataset:\n", df.shape)
print("\n3. Nama kolom:\n", df.columns, "\n\n4. Informasi dataset:\n")
print(df.info())
print("\n5. Jumlah missing values per kolom:\n", df.isnull().sum())
```

## **📌 Soal 2: Ubah Kolom Waktu & Buat Fitur Durasi**

Kolom `pickup` dan `dropoff` masih bertipe string.  
Ubahlah keduanya menjadi datetime, lalu buat kolom baru bernama `trip_duration_min` yang berisi durasi perjalanan dalam **menit**.

Setelah itu:
1. tampilkan 5 baris pertama dari kolom `pickup`, `dropoff`, dan `trip_duration_min`
2. tampilkan rata-rata durasi perjalanan


```python
# Tulis jawabanmu di sini
df['pickup'] = pd.to_datetime(df['pickup'], errors='coerce')
df['dropoff'] = pd.to_datetime(df['dropoff'], errors='coerce')
df['trip_duration_min'] = df['dropoff'] - df['pickup']
avg_duration = df['trip_duration_min'].mean()
print("1. Hasil ubah ke date:\n", df[['pickup', 'dropoff']].dtypes)
print("\n2. 5 Baris pertama:\n", df.head())
print("\n3. Rata-rata durasi perjalanan:\n", avg_duration)
```

## **📌 Soal 3: Seleksi Data dengan `loc`, `iloc`, dan Boolean Mask**

Kerjakan hal berikut:
1. Ambil kolom `fare`, `tip`, dan `total` untuk 10 baris pertama menggunakan `.loc`
2. Ambil 5 baris pertama dan 4 kolom pertama menggunakan `.iloc`
3. Tampilkan semua trip dengan:
   - `payment == "cash"`
   - `distance > 3`
   - `tip == 0`

Tampilkan 10 baris pertama dari hasil filter pada poin (3).


```python
# Tulis jawabanmu di sini
data_satu = df[['fare', 'tip', 'total']].loc[:9]
data_dua = df.iloc[:5, :4]
bool_satu = df['payment'] == 'cash'
bool_dua = df['distance'] > 3
bool_tiga = df['tip'] == 0
bool_final = bool_satu & bool_dua & bool_tiga
print("1. Ambil kolom fare, tip, dan total untuk 10 baris pertama:\n", data_satu)
print("\n2. Ambil 5 baris pertama dan 4 kolom pertama:\n", data_dua)
print("\n3. Ambil semua baris dengan metode pembayaran cash, jarak lebih dari 3, dan tip 0:\n", df[bool_final].head(10))
```

## **📌 Soal 4: Nilai Unik & Distribusi Kategori**

Tampilkan:
1. Semua nilai unik pada kolom `payment`
2. Jumlah nilai unik pada kolom `pickup_borough`
3. Banyaknya data pada setiap kategori `payment` menggunakan `value_counts()`

Lalu jawab dengan komentar: metode pembayaran apa yang paling sering muncul?


```python
# Tulis jawabanmu di sini
print("1. Semua nilai unik pada kolom payment:\n", df['payment'].unique())
print("\n2. Jumlah nilai unik pada kolom pickup_borough:\n", df['pickup_borough'].nunique())
print("\n3. Banyaknya data pada setiap kategori payment menggunakan value_counts:\n", df['payment'].value_counts())

# Metode pembayaran paling sering muncul: Credit Card dengan total 4577
```

## **📌 Soal 5: Missing Value**

Kerjakan langkah berikut:
1. Tampilkan jumlah missing value per kolom
2. Buat salinan dataset bernama `df_clean`
3. Pada `df_clean`, isi missing value:
   - kolom `payment` dengan string `"unknown"`
   - kolom `dropoff_borough` dengan string `"Unknown"`
4. Tampilkan lagi jumlah missing value dari dua kolom tersebut setelah diisi

**Catatan:** gunakan `.copy()` supaya dataset asli `df` tetap aman.


```python
# Tulis jawabanmu di sini
print("1. Tampilkan jumlah missing value per kolom:\n", df.isna().sum())
df_clean = df.copy()
print("\n2. Buat salinan dataset bernama df_clean:\n", df_clean)
df_clean['payment'].fillna('unknown')
df_clean['dropoff_borough'].fillna('Unknown')
print("\n3. Hasil isi missing value:\n", df_clean[(df_clean['payment'] == 'unknown') & (df_clean['dropoff_borough'] == 'Unknown')].head())
print("\n4. Jumlah missing value setelah diisi:\n", df_clean.isna().sum())
```

## **📌 Soal 6: Data Duplikat**

Pada dataset `taxis`, data duplikat alami jumlahnya sedikit / bahkan bisa tidak ada.  
Supaya bisa latihan, buat dataset baru bernama `df_dupes` dengan cara menambahkan **15 baris acak dari `df_clean`** ke bagian bawah dataset.

Lalu:
1. Hitung jumlah baris duplikat pada `df_dupes`
2. Buang baris duplikat tersebut dan simpan ke `df_nodup`
3. Tampilkan shape sebelum dan sesudah menghapus duplikat


```python
# Tulis jawabanmu di sini
df_dupes = pd.concat([df_clean, df_clean.sample(n=15, random_state=42)], ignore_index=True)
print("1. Jumlah baris duplikat pada df_dupes:\n", df_dupes.duplicated().sum())
df_nodupes = df_dupes.drop_duplicates().copy()
print("\n2. Buang baris ke df_nodup:\n", df_nodupes)
print("\n3. Jumlah baris sebelum duplikat dibuang:\n", df_dupes.shape[0],"\nsetelah duplikat dibuang:\n", df_nodupes.shape[0])
```

## **📌 Soal 7: Feature Engineering Sederhana**

Gunakan `df_nodup`, lalu buat:
1. Kolom `tip_pct` dengan rumus:
 
$$\text{tip_pct} = \frac{tip}{fare} \times 100$$

2. Kolom `is_long_trip`:
   - `"yes"` jika `distance >= 5`
   - `"no"` jika `distance < 5`

3. Kolom `pickup_hour` dari waktu `pickup`

Tampilkan 10 baris pertama dari kolom:
`fare`, `tip`, `tip_pct`, `distance`, `is_long_trip`, `pickup_hour`


```python
# Tulis jawabanmu di sini
df_nodupes['tip_pct'] = df_nodupes['tip'] / df_nodupes['fare'] * 100
def long_trip(duration):
    if duration >= 5:
        return 'yes'
    else:
        return 'no'
df_nodupes['is_long_trip'] = df_nodupes['distance'].apply(long_trip)
df_nodupes['pickup_hour'] = df_nodupes['pickup'].dt.hour
print("1. Kolom tip_pct:\n", df_nodupes['tip_pct'].head())
print("\n2. Kolom is_long_trip:\n", df_nodupes['is_long_trip'].head())
print("\n3. Kolom pickup_hour:\n", df_nodupes['pickup_hour'].head())
df_nodupes[['fare','tip','tip_pct','distance','is_long_trip','pickup_hour']].head(10)
```

## **📌 Soal 8: GroupBy & Aggregation**

Gunakan `df_nodup`, lalu buat ringkasan per `payment` yang berisi:
1. jumlah trip
2. rata-rata `fare`
3. rata-rata `tip`
4. rata-rata `trip_duration_min`

Urutkan hasilnya berdasarkan jumlah trip dari yang terbesar ke yang terkecil.


```python
# Tulis jawabanmu di sini
jlh_trip = df_nodupes.groupby('payment')['distance'].size()
rata_fare = df_nodupes.groupby('payment')['fare'].mean()
rata_tip = df_nodupes.groupby('payment')['tip'].mean()
rata_trip_duration = df_nodupes.groupby('payment')['trip_duration_min'].mean()
order = jlh_trip.sort_values(ascending=False).index
print("1. Total trip per metode pembayaran:\n", jlh_trip.loc[order])
print("\n2. Rata-rata fare per metode pembayaran:\n", rata_fare.loc[order])
print("\n3. Rata-rata tip per metode pembayaran:\n", rata_tip.loc[order])
print("\n4. Rata-rata durasi perjalanan per metode pembayaran:\n", rata_trip_duration.loc[order])
```

## **📌 Soal 9: GroupBy + `transform()`**

Masih menggunakan `df_nodup`, buat kolom baru bernama `fare_vs_payment_mean` dengan rumus:

\[
fare\_vs\_payment\_mean = fare - \text{rata-rata fare pada grup payment yang sama}
\]

Artinya, setiap baris dibandingkan terhadap rata-rata `fare` dari metode pembayaran yang sama.

Tampilkan 10 baris pertama dari kolom:
`payment`, `fare`, `fare_vs_payment_mean`


```python
# Tulis jawabanmu di sini
rata = df_nodupes.groupby('payment')['fare'].transform('mean')
df_nodupes['fare_vs_payment_mean'] = df_nodupes['fare'] / rata
df_nodupes[['payment', 'fare', 'fare_vs_payment_mean']].head(10)
```

## **📌 Soal 10: Sorting & Menyimpan Hasil**

Buat dataset akhir bernama `final_df` dari `df_nodup` dengan langkah:
1. Urutkan berdasarkan `total` dari terbesar ke terkecil
2. Jika nilai `total` sama, urutkan `trip_duration_min` dari terkecil ke terbesar
3. Simpan hasilnya ke file bernama **`taxis_final.csv`**

Tampilkan 10 baris pertama dari `final_df`.


```python
# Tulis jawabanmu di sini
final_df = df_nodupes.copy()
print("1. Hasil urutkan berdasarkan total dari terbesar:\n", final_df[['total']].sort_values(by='total', ascending=False).head(10))
print("\n2. Hasil double grouping:\n", final_df[['total','trip_duration_min']].sort_values(by=['total', 'trip_duration_min'], ascending=[False, True]).head(10))
final_df.to_csv('taxis_final.csv', index=False)
```
